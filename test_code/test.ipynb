```python
import numpy as np

g = 9.80665 # m/s^2
m = 1.238 # kg
l = 0.241671 # m
I = m*l*l # kg.m^2
wn = np.sqrt(g/l)
fn = 1/(2*np.pi)*np.sqrt(g/l)
print(wn, fn)
```

```python
import matplotlib.pyplot as plt

theta_0 = np.pi/2
omega_0 = 0
time = np.linspace(0, 10, 1000)
theta_t = theta_0*np.cos(time*wn)

fig = plt.figure(figsize=(10,5))
plt.plot(time, theta_t*180/(2*np.pi))
plt.grid(True)
plt.xlabel('time($t$)')
plt.ylabel('angle($deg$)')
plt.show()
```
